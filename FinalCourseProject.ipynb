```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
df=pd.read_csv('data.csv')
df1=df.head(10)
chart=plt.figure()
X=chart.add_subplot(1,1,1)
y=np.arange(len(df1.Country))
X.set_xlabel('Overall Happiness Score')
X.set_ylabel('Top 10 Countries')
plt.xlim(0,8)
X.set_title('Overall Happiness Score of the top 10 countries')
X.barh(df1['Country'],df1['Economy'],color='r')
X.barh(df1['Country'],df1['Family'],color='g',left= df1['Economy'])
X.barh(df1['Country'],df1['Health'],color='b',left= df1['Economy']+df1['Family'])
X.barh(df1['Country'],df1['Freedom'],color='c',left= df1['Economy']+df1['Family']+df1['Health'])
X.barh(df1['Country'],df1['Generosity'],color='m',left= df1['Economy']+df1['Family']+df1['Health']+df1['Freedom'])
X.barh(df1['Country'],df1['Corruption'],color='y',left= df1['Economy']+df1['Family']+df1['Health']+df1['Freedom']+df1['Generosity'])
X.barh(df1['Country'],df1['Dystopia'],color='k',left= df1['Economy']+df1['Family']+df1['Health']+df1['Freedom']+df1['Generosity']+df1['Corruption'])
plt.legend(df1.columns[3:10])
plt.show()
```

```python
df2=df[df['Region']=='Africa']
chart=plt.figure()
X=chart.add_subplot(1,1,1)
y=np.arange(len(df2.Country))
X.set_xlabel('Overall Happiness Score')
X.set_ylabel('Countries in Africa')
plt.xlim(0,8)
X.set_title('Overall Happiness Score of the top 10 countries')
X.barh(df2['Country'],df2['Economy'],color='r')
X.barh(df2['Country'],df2['Family'],color='g',left= df2['Economy'])
X.barh(df2['Country'],df2['Health'],color='b',left= df2['Economy']+df2['Family'])
X.barh(df2['Country'],df2['Freedom'],color='c',left= df2['Economy']+df2['Family']+df2['Health'])
X.barh(df2['Country'],df2['Generosity'],color='m',left= df2['Economy']+df2['Family']+df2['Health']+df2['Freedom'])
X.barh(df2['Country'],df2['Corruption'],color='y',left= df2['Economy']+df2['Family']+df2['Health']+df2['Freedom']+df2['Generosity'])
X.barh(df2['Country'],df2['Dystopia'],color='k',left= df2['Economy']+df2['Family']+df2['Health']+df2['Freedom']+df2['Generosity']+df2['Corruption'])
plt.legend(df2.columns[3:10])
plt.show()
```
